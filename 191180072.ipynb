```python
#Q18


Survived_passengers=df_titanic[df_titanic["Survived"]==1]
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10, 4))

ax = sns.histplot(Survived_passengers[Survived_passengers['Sex']=="female"].Age, bins=10, label = 'female', ax = axes[0],color='b')
ax = sns.histplot(Survived_passengers[Survived_passengers['Sex']=="male"].Age, bins=10, label = 'male', ax = axes[0], color='r')
ax.legend()

ax.set_title('age distribution of survived')

#Looking at the histogram, it is seen that the 
#survival rate of female passengers is higher.


#This means that women have a higher chance of survival


```
